```python
#DataReduction(Sampling)
#Date:4/6/2023
#Author:Kevin ZHENG
#This notebook samples 10000 flights to reduce the size of dataset using two methods. 
#The output file with "RAW" in name means it is 10000 out of 86866 flights.
#The output file with "ARRDELAY" in name means it is 10000 out of 19252 arrival-delayed flights, which have detailed information of delay.
```

```python
import pandas as pd
import seaborn as sns
import glob
import os
```

```python
#Read the dataset
df=pd.read_csv("E:\EE461P\Data\AUS_Classification.csv",index_col=0)
df.info()
df.head()
```

```python
df.loc[df['ARR_DELAY_GROUP']>=1]
```

```python
#Filter the delayed filghts(arrival)
df_d=df.loc[df['ARR_DELAY_GROUP']>=1]
```

```python
#Random sampling
df_d_reduction=df_d.sample(n=10000,random_state=42)
df_d_reduction.info()
df_d_reduction.head()
```

```python
df_d_reduction.to_csv('E:\EE461P\Data\AUS_10K_ARRDELAY.csv')
```

```python
df_r_reduction=df.sample(n=10000,random_state=42)
df_r_reduction= df_r_reduction.reset_index(drop=True)
df_r_reduction.info()
df_r_reduction.head()
```

```python
df_r_reduction.to_csv('E:\EE461P\Data\AUS_10K_RAW.csv')
```

```python
df2=pd.read_csv("E:\EE461P\Data\AUS_Regression.csv",index_col=0)
df2.info()
df2.head()
```

```python
df2_d=df2.loc[df2['ARR_DELAY']>=15]
```

```python
df2_d_reduction=df2_d.sample(n=10000,random_state=42)
df2_d_reduction.info()
df2_d_reduction.head()
```

```python
df2_d_reduction.to_csv('E:\EE461P\Data\AUS_10K_ARRDELAY_Regression.csv')
```

```python
df2_r_reduction=df2.sample(n=10000,random_state=42)
df2_r_reduction= df2_r_reduction.reset_index(drop=True)
df2_r_reduction.info()
df2_r_reduction.head()
```

```python
df2_r_reduction.to_csv('E:\EE461P\Data\AUS_10K_RAW_Regression.csv')
```
