```python
# importing libraries
import pandas as pd
import glob
import os

# merging the files
joined_files = os.path.join("E:\EE461P\Data", "*.csv")
joined_list = glob.glob(joined_files)
joined_list
```

```python
#Read the files
df = pd.concat(map(pd.read_csv, joined_list), ignore_index=True)
df.head()
```

```python
df.info()
```

Note:

All **NaN** values are replaced with **0**.

All **CRS** values are "scheduled" values.

**DEP_DELAY** / **ARR_DELAY** contain **negative** values, which means that the flight departures / arrives **earlier** than scheduled.

**DEL_15** values are set to 1 if delay >= 15, otherwise 0

|Code| Description|
|--- | ---|
|0| Delay < 15 minutes|
|1| Delay >= 15 minutes |

**DEL_GROUP** values are based on the following rules:

|Code|	Description|
| --- | --- |
| -2	| Delay < -15 minutes |
|-1	|Delay between -15 and -1 minutes|
|0	|Delay between 0 and 14 minutes|
|1	|Delay between 15 to 29 minutes|
|2	|Delay between 30 to 44 minutes|
|3	|Delay between 45 to 59 minutes|
|4	|Delay between 60 to 74 minutes|
|5	|Delay between 75 to 89 minutes|
|6	|Delay between 90 to 104 minutes|
|7	|Delay between 105 to 119 minutes|
|8	|Delay between 120 to 134 minutes|
|9	|Delay between 135 to 149 minutes|
|10	|Delay between 150 to 164 minutes|
|11	|Delay between 165 to 179 minutes|
|12	|Delay >= 180 minutes|

**DISTANCE_GROUP** is quite similar,

|Code|	Description|
|---|---|
|1	|Less Than 250 Miles|
|2	|250-499 Miles|
|3	|500-749 Miles|
|4	|750-999 Miles|
|5	|1000-1249 Miles|
|6	|1250-1499 Miles|
|7	|1500-1749 Miles|
|8	|1750-1999 Miles|
|9	|2000-2249 Miles|
|10	|2250-2499 Miles|
|11	|2500 Miles and Greater|


```python
#Delete cancelled or diverted flights
df= df.loc[df['CANCELLED'] == 0]
df= df.loc[df['DIVERTED'] == 0]
```

```python
#Drop unused status
df = df.drop(columns=['CANCELLED'])
df = df.drop(columns=['DIVERTED'])
#Drop the duplicated columns
df = df.drop(columns=['DEP_DELAY_NEW','ARR_DELAY_NEW'])

#Fill all empty cells with 0
df.fillna(0,inplace=True)
```

```python
#Select AUS, DFW and IAH
df_AUS=df.loc[df['ORIGIN_AIRPORT_SEQ_ID'] == 1042302 ]
df_DFW=df.loc[df['ORIGIN_AIRPORT_SEQ_ID'].between(1129801,1129806)]
df_IAH=df.loc[df['ORIGIN_AIRPORT_SEQ_ID'].between(1226602,1226603)]

#Three largest Texas airports combined.
frames = [df_AUS,df_DFW,df_IAH]
df_TX=pd.concat(frames)
df_TX= df_TX.reset_index(drop=True)
```

```python
#Dataframe for regression. Drop all grouping columns.
df_TX_R=df_TX
df_TX_R=df_TX_R.drop(columns=['DEP_DEL15','DEP_DELAY_GROUP','ARR_DEL15','ARR_DELAY_GROUP','DISTANCE_GROUP'])
df_TX_R.info()
```

```python
#Dataframe for classification, all grouping codes are reserved.
df_TX_C=df_TX
df_TX_C.info()
```

```python
df_TX_C.to_csv('E:\EE461P\Data\TX_Classification.csv')
```

```python
df_TX_R.to_csv('E:\EE461P\Data\TX_Regression.csv')
```
